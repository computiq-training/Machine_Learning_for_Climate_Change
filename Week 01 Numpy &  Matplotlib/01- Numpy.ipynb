# The Basics
NumPy’s main object is the homogeneous multidimensional array. It is a table of elements (usually numbers), all of the same type, indexed by a tuple of non-negative integers. In NumPy dimensions are called axes.

For example, the array for the coordinates of a point in 3D space, `[1, 2, 1]`, has one axis. That axis has 3 elements in it, so we say it has a length of 3. In the example pictured below, the array has 2 axes. 

The first axis has a length of 2, the second axis has a length of 3.




```
[[1., 0., 0.],
 [0., 1., 2.]]

```

NumPy’s array class is called `ndarray`. It is also known by the alias array. 

Note that `numpy.array` is not the same as the Standard Python Library class `array.array`, which only handles one-dimensional arrays and offers less functionality. The more important attributes of an `ndarray` object are:

###`ndarray.ndim`

the number of axes (dimensions) of the array.

###`ndarray.shape`

the dimensions of the array. This is a tuple of integers indicating the size of the array in each dimension. For a matrix with n rows and m columns, shape will be (n,m). The length of the shape tuple is therefore the number of axes, ndim.

###`ndarray.size`

the total number of elements of the array. This is equal to the product of the elements of shape.

### `ndarray.dtype`
an object describing the type of the elements in the array. One can create or specify dtype’s using standard Python types. Additionally NumPy provides types of its own. `numpy.int32`, `numpy.int16`, and `numpy.float64` are some examples.



````python
# define function to quickly show the different properties of a Numpy array
def describe(arr):
    """Print out a summary of an array."""
    print('#'*100)
    print(f"values: \n{arr}")
    print(f"dimensions: {arr.ndim}")
    print(f"shape: {arr.shape}")
    print(f"size: {arr.size}")
    print(f"dtype: {arr.dtype}")
    print('#'*100)
````

### General Ndarray Creation Functions


The ndarray creation functions e.g. `numpy.ones`, `numpy.zeros`, and `random` define arrays based upon the desired shape. The ndarray creation functions can create arrays with any dimension by specifying how many dimensions and length along that dimension in a `tuple` or `list`.

`numpy.zeros` will create an array filled with 0 values with the specified shape. The default dtype is float64:



````python
arr = np.zeros((3, 5))  # create a 3x5 array of zeros
describe(arr)

````

````python
arr = np.zeros((3, 5),dtype=np.int32)  # create a 3x5 array of zeros with dtype of int32
describe(arr)
````

`np.zeros_like` return an array of zeros with the same shape and type as a given array.



````python
line1 = [1, 2, 3]
line2 = [4, 5, 6]

arr_1 = np.array([line1, 
                line2])  # create a 2 x 3 array from the two lines 

arr = np.zeros_like(arr_1)  # creat an array of ones have the same shape of "arr_1"
describe(arr)
````

`numpy.ones` will create an array filled with 1 values. It is identical to zeros in all other respects as such:



````python
arr = np.ones((2, 2))  # create a 2x2 array of ones
describe(arr)
````

````python
line1 = [1, 2, 3]
line2 = [4, 5, 6]

arr_1 = np.array([line1, 
                  line2])  # create a 2 x 3 array from the two lines 

arr = np.ones_like(arr_1)  # creat an array of ones have the same shape of "arr_1"
describe(arr)
````

`np.full` Return a new array of given shape and type, filled with `fill_value`.



````python
arr = np.full(shape=(2, 2), fill_value=10)
describe(arr)
````

### Generate Random Number
`NumPy` offers the `random` module to work with random numbers.



#### Random Integers Numbers

`np.random.randint` we using it to generate ndarray with random integer. 


**Parameters:**

`low`: lowint or array-like of ints
Lowest (signed) integers to be drawn from the distribution (unless high=None, in which case this parameter is one above the highest such integer).

`high`: or array-like of ints, optional
If provided, one above the largest (signed) integer to be drawn from the distribution (see above for behavior if high=None). If array-like, must contain integer values

`size`: sizeint or tuple of ints, optional
Output shape. If the given shape is, e.g., (m, n, k), then m * n * k samples are drawn. Default is None, in which case a single value is returned.

`dtype`: dtypedtype, optional
Desired dtype of the result. Byteorder must be native. The default value is int.

````python
arr = np.arange(100) # create vector with 100 element 
print('Befor reshaping')
describe(arr)
# Resahpe the numpy array from shape =(100,) to shape =(20,5)
arr = arr.reshape(20,5)
print('after reshaping')
describe(arr)
````

### Dot Product

@ or np.dot or np.matmul those functions returns the dot product of two arrays. For 2-D vectors, it is the equivalent to matrix multiplication. For 1-D arrays, it is the inner product of the vectors. For N-dimensional arrays, it is a sum product over the last axis of a and the second-last axis of b.

````python
data_1 = np.array([[1,2,3],
                   [3,2,1]])

data_2 = np.array([[1,2],
                   [3,2],
                   [2,7]])


arr = data_1 @ data_2
describe(arr)
````

**OR**

````python
arr = np.matmul(data_1 , data_2)
describe(arr)
````

In a multidimensional array, `...` can be used to stand for multiple instances of `:`, enough to form a complete index in an expression. 

````python
data = np.arange(27).reshape(3, 3, 3)  # create a 3x3x3 array with ints 0-26

print("-" * 40)
print(data)
print("-" * 40)
print("first: ")
print(data[1, ...])  # same as data[1,:,:]

print("-" * 40)
print("second: ")
print(data[1, 1, ...])  # same as data[1,1,:]
print("-" * 40)
print("third: ")
print(data[..., 2])  # same as data[:,:,2]
````

# Numpy Array Stacking

Arrays can also be combined, or "stacked".

we use `vstack` to stack arrays in sequence vertically (row wise)


we use `hstck` to stack arrays in sequence horizontally (column wise)

````python
a = np.array([1, 2, 3, 4, 5])
b = np.array([6, 7, 8, 9, 10])
c = np.array([11, 12, 13, 14, 15])

````

````python
d = np.vstack((a, b, c))  # Stack arrays in sequence vertically (row wise)
describe(d)
````

````python
e = np.hstack((a, b, c))  # Stack arrays in sequence horizontally (column wise)
describe(e)
````
